```python
#Chef and dolls
t = int(input())
for i in range(t):
  n = int(input())
  miss = 0
  for j in range(n):
    dolls = int(input())
    miss ^= dolls
  print("Output:",miss)
```

```python
#Aggressive Cows
def is_possible(stalls, cows, mid_dist):
    count = 1
    last_placed_pos = stalls[0]

    for i in range(1, len(stalls)):
        if stalls[i] - last_placed_pos >= mid_dist:
            count += 1
            last_placed_pos = stalls[i]
            if count >= cows:
                return True
    return False

def solve():
    n, c = map(int, input().split())
    stalls = [int(input()) for _ in range(n)]

    stalls.sort()

    low = 1
    high = stalls[-1] - stalls[0]
    ans = 0

    while low <= high:
        mid = (low + high) // 2

        if is_possible(stalls, c, mid):
            ans = mid
            low = mid + 1
        else:
            high = mid - 1

    print(ans)

if __name__ == "__main__":
    t = int(input())
    for _ in range(t):
        solve()
```
